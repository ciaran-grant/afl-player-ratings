Stats per Score

```python
def stats_per_point(player_stats, stat_cols):
    
    cols = stat_cols + ['Behinds', 'Goals']
    stats = player_stats[cols].sum()
    stats['Points'] = stats['Behinds'] + stats['Goals']*6
    
    for stat in stat_cols:
        stats["Points_per_"+stat] = stats['Points'] / stats[stat]
    
    df_stats = pd.DataFrame(stats, columns = ['Stat'])
    df_stats['Stat'] = df_stats['Stat'].map("{:.2f}".format)
    return df_stats
    
```

```python
def team_stats_per_point(player_stats):
    
    team_stats = player_stats.groupby('Team').agg(
        Contested_Possessions = ("Contested_Possessions", "sum"),
        Offensive_Contest_Wins = ("Offensive_Contest_Wins", "sum"),
        Defensive_Contest_Losses = ("Defensive_Contest_Losses", "sum"),
        Defensive_Half_Pressure_Acts = ("Defensive_Half_Pressure_Acts", "sum"),
        Disposals = ("Disposals", "sum"),
        AFL_Fantasy_Points = ("AFL_Fantasy_Points", "sum"),
        Effective_Disposals = ("Effective_Disposals", "sum"),
        Effective_Kicks = ("Effective_Kicks", "sum"),
        Frees_Against = ("Frees_Against", "sum"),
        Frees_For = ("Frees_For", "sum"),
        Ground_Ball_Gets = ("Ground_Ball_Gets", "sum"),
        Handballs = ("Handballs", "sum"),
        Hit_Outs = ("Hit_Outs", "sum"),
        Hit_Outs_To_Advantage = ("Hit_Outs_To_Advantage", "sum"),
        Inside_50s = ("Inside_50s", "sum"),
        Intercept_Marks = ("Intercept_Marks", "sum"),
        Intercepts = ("Intercepts", "sum"),
        Kicks = ("Kicks", "sum"),
        Marks = ("Marks", "sum"),
        Marks_Inside_50 = ("Marks_Inside_50", "sum"),
        Marks_On_Lead = ("Marks_On_Lead", "sum"),
        Metres_Gained = ("Metres_Gained", "sum"),
        Pressure_Acts = ("Pressure_Acts", "sum"),
        Player_Rating_Points = ("Player_Rating_Points", "sum"),
        Ruck_Contests = ("Ruck_Contests", "sum"),
        Shots_At_Goal = ("Shots_At_Goal", "sum"),
        Spoils = ("Spoils", "sum"),
        Stoppage_Clearances = ("Stoppage_Clearances", "sum"),
        Tackles = ("Tackles", "sum"),
        Tackles_Inside_50 = ("Tackles_Inside_50", "sum"),
        Clearances = ("Clearances", "sum"),
        Possessions = ("Possessions", "sum"),
        Uncontested_Possessions = ("Uncontested_Possessions", "sum"),
        Behinds = ("Behinds", "sum"),
        Goals = ("Goals", "sum")
    )
    team_stats['Points'] = team_stats['Behinds'] + team_stats['Goals']*6
    
    stat_cols = [x for x in team_stats if x not in ['Behinds', 'Goals', 'Points']]
    for stat in stat_cols:
        team_stats["Points_per_"+stat] = team_stats['Points'] / team_stats[stat]
    
    return team_stats
    
```

```python
def sort_team_summary(team_summary, stat):
    
    return team_summary[[stat, 'Points_per_'+stat, 'Points']].sort_values('Points_per_'+stat, ascending=False)
```

```python
team_summary = team_stats_per_point(player_stats)
sort_team_summary(team_summary, "Intercept_Marks")
```
