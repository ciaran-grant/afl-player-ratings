```python
import pandas as pd
import re
import numpy as np

import os
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
exp_vaep_data = pd.read_csv("/Users/ciaran/Documents/Projects/AFL/git-repositories/afl-player-ratings/exp_vaep/data/exp_vaep_values_v2.csv")
exp_vaep_data['season'] = exp_vaep_data['match_id'].apply(lambda x: int(x[:4]))
exp_vaep_data['player_season'] = exp_vaep_data['player'] + "_" + exp_vaep_data['season'].astype(str)

exp_vaep_data['home_team'] = exp_vaep_data['match_id'].apply(lambda x: x.split("_")[1]).str.replace(r"([A-Z])", r" \1").str.strip()
exp_vaep_data['away_team'] = exp_vaep_data['match_id'].apply(lambda x: x.split("_")[-1]).str.replace(r"([A-Z])", r" \1").str.strip()
exp_vaep_data['opponent'] = np.where(exp_vaep_data['team'] == exp_vaep_data['home_team'], exp_vaep_data['away_team'], exp_vaep_data['home_team'])
exp_vaep_data.head()
```
