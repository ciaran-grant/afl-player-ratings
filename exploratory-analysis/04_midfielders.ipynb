```python
import pandas as pd
import numpy as np

import os
import warnings
warnings.filterwarnings('ignore')
pd.options.display.max_rows = 999
pd.options.display.max_columns = 999
```

```python
player_stats = pd.read_csv("/Users/ciaran/Documents/Projects/AFL/git-repositories/afl-player-ratings/data/player_stats.csv")
player_stats['Season'] = player_stats['Match_ID'].apply(lambda x: int(x[:4]))
player_stats['Score'] = player_stats['Behinds'] + player_stats['Goals']*6
player_stats.head()
```

```python
midfield = player_stats[player_stats['Position_Group'] == "Midfield"]
```

```python
midfield['Position_Sub_Group'].value_counts()
```

```python
midfield_2021 = midfield[midfield['Season'] == 2021]
midfield_2022 = midfield[midfield['Season'] == 2022]
midfield_2023 = midfield[midfield['Season'] == 2023]
```

```python
def rank_players(player_stats, stat, top_n = 10):
    
    return player_stats.groupby(["Season", 'Player', 'Team']).sum()[stat].reset_index().sort_values(by=stat, ascending=False).head(top_n)
```
